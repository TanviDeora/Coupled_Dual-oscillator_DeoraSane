```python
import numpy as np
import pandas as pd
import glob
import os

from scipy.io import loadmat
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
```

## Slit Sc

```python
direc = r"../dataFolder/AmplitudeAnalysis/v1/slit_sc/"
```

```python
df = pd.read_csv(direc + 'median_summary_edited.csv')
df.head()
```

```python
# do wilcoxon for each winglength
slit_sc_SumStats = pd.DataFrame(data = None,
                                index = df.binned.unique(),
                                columns = ['wilcoxon', 'kstest', 'n1', 'n2'])

for binned, df_binned in df.groupby('binned'):
#     print(binned)
    clip = df_binned.loc[df_binned.bodypart == 'clip', 'med_amplitude']
    n1 = len(clip)
    intact = df_binned.loc[df_binned.bodypart == 'intact', 'med_amplitude']
    n2 = len(clip)
    
    U, p = stats.wilcoxon(clip, intact, alternative='two-sided')
    u_ks, p_ks = stats.ks_2samp(clip, intact)
    slit_sc_SumStats.loc[binned, 'wilcoxon']= p
    slit_sc_SumStats.loc[binned, 'kstest'] = u_ks
    slit_sc_SumStats.loc[binned, ['n1','n2']] = n1, n2
slit_sc_SumStats
```
